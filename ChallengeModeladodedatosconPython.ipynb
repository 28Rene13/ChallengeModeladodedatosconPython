### Importación de datos



```python
# Importación de librerías a utilizar.
import pandas as pd #Para la manipulación y el análisis de datos
import matplotlib.pyplot as plt #Importar el módulo pyplot de matplotlib para crear gráficos.

#Impotación de los datos para realizar el analisis.
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

#1. Análisis de facturación



```python
# Se calcula la Facturación por tienda y se muestra.

# Lista de tiendas
facturacion_tienda1 = tienda1['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print(f"La Tienda 1 facturó: ${facturacion_tienda1:,.2f}")
print(f"La Tienda 2 facturó: ${facturacion_tienda2:,.2f}")
print(f"La Tienda 3 facturó: ${facturacion_tienda3:,.2f}")
print(f"La Tienda 4 facturó: ${facturacion_tienda4:,.2f}")
```

```python
# Se crea un diccionario con los totales.
facturaciones = { 'Tienda 1': facturacion_tienda1,
                  'Tienda 2': facturacion_tienda2,
                  'Tienda 3': facturacion_tienda3,
                  'Tienda 4': facturacion_tienda4}

#Se calcula y muestra la facturacion total.
total_general = sum(facturaciones.values())
print(f"Total general facturado por todas las tiendas: ${total_general:,.2f}")

#Se determina y muestra la tienda que mas facturo.
tienda_max = max(facturaciones, key=facturaciones.get)
print(f"La tienda con mayor facturación es {tienda_max} con ${facturaciones[tienda_max]:,.2f}")

#Se determina y muestra la tienda que menos facturo.
tienda_min = min(facturaciones, key=facturaciones.get)
print(f"La tienda con menor facturación es {tienda_min} con ${facturaciones[tienda_min]:,.2f}")
```

```python
# Se grafica la facturación de las tiendas.

# Lista de nombres de tiendas.
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Lista de valores de facturación.
valores = [facturaciones[tienda] for tienda in tiendas]

# Crea la gráfica de barras.
barras = plt.bar(tiendas, valores, color='green')
plt.title('Facturación por tienda')
plt.xlabel('Tiendas')

plt.gca().axes.get_yaxis().set_visible(False) # Ocultar el eje Y.

# Agrega la facturación de cada tienda encima de su grafica.
for barra, valor in zip(barras, valores):
    plt.text(barra.get_x() + barra.get_width()/2,
             barra.get_height(),
             f"${valor:,.2f}",
             ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout() # Ajusta automáticamente el diseño de la grafica.
plt.show() # Muestra el gráfico.
```

# 2. Ventas por categoría

```python
# Se crea una lista de tuplas, donde cada tupla contiene el número de la tienda y su respectivo DataFrame
tts = [
    (1, tienda1),
    (2, tienda2),
    (3, tienda3),
    (4, tienda4)
]

#Se determina y muestra el número de ventas de cada tienda por cada categoría de productos.
for t in tts:
    print(f"\n📦 Tienda {t[0]} - Ventas por Categoría:\n")

    ventas_por_categoria = (
        t[1].groupby(['Categoría del Producto'])
            .size()
            .sort_values(ascending=False)
            .reset_index(name='Número de Ventas')
    )

    print(ventas_por_categoria)

# Agregar un separador visual entre cada tienda.
    print("\n" + "-" * 50)  # Línea de separación.
```

# 3. Calificación promedio de la tienda


```python
# Se calcula el promedio y se muestra de la columna "Calificación" para cada tienda.

# Diccionario con los nombres de cada tienda.
tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Diccionario para guardar los promedios.
promedios = {}

# Calcular promedio de calificación por tienda.
for nombre, df in tiendas.items():
    promedio = round(df["Calificación"].mean(), 2)
    promedios[nombre] = promedio
    print(f"Calificación promedio de la {nombre}: {promedio}")
```

```python
#Se grafica la Calificación de las tiendas.

# Extraemos nombres y valores del diccionario
nombres_tiendas = list(promedios.keys())
valores_promedios = list(promedios.values())

# Crear gráfico de barras
plt.bar(nombres_tiendas, valores_promedios, color='red')

# Etiquetas y título
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Tiendas")
plt.gca().axes.get_yaxis().set_visible(False) # Ocultar el eje Y.

# Mostrar los valores encima de cada barra.
for i, valor in enumerate(valores_promedios):
    plt.text(i, valor + 0.05, f"{valor:.2f}", ha='center')

plt.tight_layout() # Ajusta automáticamente el diseño de la grafica.
plt.show() # Muestra la grafica.
```

# 4. Productos más y menos vendidos

```python
#Se determina el producto más y menos ventido de cada tienda y se grafica la venta de sus productos.

# Diccionario con los nombres de cada tienda.
tiendas = {
    "Tienda 1": tienda1,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Colores a utilizar en las graficas.
colores_mas_vendidos = ['#4caf50', '#ffeb3b', '#00bcd4', '#ff5722', '#2196f3']  # Verde, amarillo, azul, naranja, azul brillante.
colores_menos_vendidos = ['#e91e63', '#f44336', '#ff9800', '#9c27b0', '#ffc107']  # Rosa brillante, rojo, naranja, morado, amarillo.

# Recorremos cada tienda.
for nombre, df in tiendas.items():
    print(f"\n📦 {nombre} - Top productos más vendidos:")
    productos_mas_vendidos = df["Producto"].value_counts().head()  # Top 5 más vendidos.
    print(productos_mas_vendidos)

    print(f"\n📉 {nombre} - Productos menos vendidos:")
    productos_menos_vendidos = df["Producto"].value_counts().tail()  # Top 5 menos vendidos.
    print(productos_menos_vendidos)

    # Agregar un separador visual entre cada tienda.
    print("\n" + "-" * 50)  # Línea de separación.

    # Gráfico circular de los más vendidos.
    plt.figure(figsize=(6, 6))
    plt.pie(
        productos_mas_vendidos,
        labels=productos_mas_vendidos.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=colores_mas_vendidos
    )
    plt.title(f"Top 5 productos más vendidos - {nombre}")
    plt.tight_layout()
    plt.show()

    # Gráfico circular de los menos vendidos.
    plt.figure(figsize=(6, 6))
    plt.pie(
        productos_menos_vendidos,
        labels=productos_menos_vendidos.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=colores_menos_vendidos
    )
    plt.title(f"Top 5 productos menos vendidos - {nombre}")
    plt.tight_layout() # Ajusta automáticamente el diseño de la grafica
    plt.show() # Muestra la grafica
```

# 5. Envío promedio por tienda

```python
# Cálcula y muestra el costo de envío promedio por tienda.
envio_prom_tienda1 = round(tienda1["Costo de envío"].mean(), 2)
envio_prom_tienda2 = round(tienda2["Costo de envío"].mean(), 2)
envio_prom_tienda3 = round(tienda3["Costo de envío"].mean(), 2)
envio_prom_tienda4 = round(tienda4["Costo de envío"].mean(), 2)


print(f"Costo de envío promedio de la Tienda 1: {envio_prom_tienda1} COP")
print(f"Costo de envío promedio de la Tienda 2: {envio_prom_tienda2} COP")
print(f"Costo de envío promedio de la Tienda 3: {envio_prom_tienda3} COP")
print(f"Costo de envío promedio de la Tienda 4: {envio_prom_tienda4} COP")

# Datos para graficar.
costos_envio = [envio_prom_tienda1, envio_prom_tienda2, envio_prom_tienda3, envio_prom_tienda4]  # Lista de costos de envío.
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]  # Nombres de las tiendas.

# Se crea el gráfico de barras.
plt.figure(figsize=(10, 7))  # Aumentar el tamaño del gráfico para más espacio.
bars = plt.bar(tiendas, costos_envio, color=["#4CAF50", "#FFEB3B", "#2196F3", "#FF5722"])  # Barras con colores diferentes.

# Se coloca los valores sobre cada barra.
for bar in bars:
    yval = bar.get_height()  # Obtener el valor de cada barra.
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 100,  # Aumentar el desplazamiento hacia arriba.
             f'{yval}', ha='center', va='bottom', fontsize=12, fontweight='bold')

# Títulos y etiquetas del gráfico
plt.title("Costo de envío promedio por tienda", fontsize=16, fontweight='bold', pad=30)  # Aumentar espacio en el título.
plt.xlabel("Tiendas", fontsize=14)  # Etiqueta eje X.
plt.ylabel("Pesos colombianos (COP)", fontsize=14)  # Etiqueta eje Y.

# Se ajusta el límite superior del eje Y para dejar un espacio.
plt.ylim(0, max(costos_envio) + 1000)  # Ampliar el límite superior para dejar más espacio.

# Se ajustan los márgenes para evitar la superposición con el borde.
plt.subplots_adjust(top=0.85, bottom=0.25)  # Más espacio arriba y abajo.


plt.tight_layout() # Ajusta automáticamente el diseño de la grafica.
plt.show() # Se muestra la grafica.
```

# **Análisis y Recomendaciones**








```python
#Se crea una funcion para mostra la facturación, promedio de calificación y costo promedio de encio de cada tienda.

def analizar_tienda(df, nombre):
    # Se verifica si hay valores nulos y manejarlo.
    if df.isnull().any().any():
        print(f"⚠️ Advertencia: Hay valores faltantes en los datos de {nombre}. Los valores nulos serán ignorados en los cálculos.")

    # Cálculos.
    facturacion = df["Precio"].sum() if df["Precio"].notnull().any() else 0
    promedio_calificacion = df["Calificación"].mean() if df["Calificación"].notnull().any() else 0
    costo_envio_promedio = df["Costo de envío"].mean() if df["Costo de envío"].notnull().any() else 0
    ventas_por_categoria = df["Categoría del Producto"].value_counts()

    # Mostrar resultados.
    print(f"\n📦 {nombre}")
    print(f"Facturación total: ${facturacion:,.2f} COP")
    print(f"Promedio de Calificación: {promedio_calificacion:.2f}")
    print(f"Costo de envío promedio: ${costo_envio_promedio:.2f}")

    # Mostrar ventas por categoría.
    print("Ventas por categoría:")
    print(ventas_por_categoria)
    print("-" * 40)

    # Se crea un gráfico de ventas por categoría.
    plt.figure(figsize=(8, 6))
    ventas_por_categoria.plot(kind='bar', color='skyblue')
    plt.title(f"Ventas por Categoría - {nombre}")
    plt.xlabel("Categoría de Producto")
    plt.ylabel("Número de Ventas")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

# Se jecuta para cada tienda a funcion.
analizar_tienda(tienda1, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")

```
